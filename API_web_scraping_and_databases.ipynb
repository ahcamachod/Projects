```python
from bs4 import BeautifulSoup
import tqdm
import numpy as np
import requests
import re
```

```python
url = 'https://www.worldometers.info/coronavirus/' # This is the url to be scraped from.
response = requests.get(url) # we use this command to get the response from the url
response.text #This command returns a string that contains the information of the url
```

```python
pattern ='"(\w{3} \d{2})"'
dates = re.findall(pattern, response.text) # Using regular expressions to locate date information
dates = dates[:70]
print(dates)
```

```python
dates1= str(dates)
pattern1 = '2020-01-'
pattern2 = '2020-02-'
pattern3 = '2020-03-'
dates2 = re.sub(('Jan '), pattern1, dates1)
dates3 = re.sub('Feb ', pattern2, dates2)
date = re.sub('Mar ', pattern3, dates3)
date = date.replace('\'', '').replace('[', '').replace(']', '').replace(' ', '')
date = date.split(',')
date = pd.to_datetime(date) # Since date was not in datetime format, we applied the "pd.to_datetime" command
date
```

### Now we do the soup to gather more information from the website:

```python
soup = BeautifulSoup(response.text, "html.parser")
soup #This command shows the html tree to find the appropriate data from the website.
```

```python
js = soup.find_all('script',{'type' : 'text/javascript'})
js
```

```python
len(js)
```

```python
#REGEX was used to find the actual number of coronavirus worldwide confirmed cases.
js1 = js[10].text
pattern = 'data: \[580.*]?'
js2 = re.findall(pattern, js1)
js3 = js2[0]
js3
```

```python
pattern = '580,8.*8355'
data = re.findall(pattern, js3)
data1 = str(data).replace("'", '').replace('[', '').replace(']', '')
data = data1.split(',')
data = pd.to_numeric(data)
len(data)
```

```python
type(data)
data
```

```python
df3 = pd.DataFrame(list(zip(date, data))) # we create another dataset with information related to COVID-19 reported cases worldwide
df3.columns = ['date', 'COVID-19 cases']
df3
```

```python
nep_vs_corona = pd.merge(df2, df3, on=['date']) # Now we merge the two dataframes to see
nep_vs_corona.head()
```
