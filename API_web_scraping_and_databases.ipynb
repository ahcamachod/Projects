```python
from datetime import datetime
import pandas as pd
```

```python
from iexfinance.stocks import Stock # This is one of the main libraries
from iexfinance.stocks import get_historical_data
nep = ['NOK','ERIC','ZTCOY','JNPR','CSCO'] #Network Equipment Provider list
df = Stock(nep, output_format='pandas', token=token) # Create a Pandas dataframe with the provided information
df1 = df.get_historical_prices(range='ytd') #Focus on the current year's stock prices of the selected companies'shares
df1.head() 
```

```python
df1.shape
```

```python
# Since we only need the closing stock prices for each company, we do some manipulation of the columns to get what we need:
df2 = df1.copy()
df2['NOK'] = df2['NOK'].loc[:,['close']]
df2['ERIC'] = df2['ERIC'].loc[:,['close']]
df2['ZTCOY'] = df2['ZTCOY'].loc[:,['close']]
df2['JNPR'] = df2['JNPR'].loc[:,['close']]
df2['CSCO'] = df2['CSCO'].loc[:,['close']]
df2 = df2[['NOK','ERIC','ZTCOY','JNPR','CSCO']].dropna(axis=1, how='all')
df2.reset_index().head()
```

```python
df2.columns = df2.columns.droplevel(1) # We use this command to remove the 'close' level in each column.
```



## 2. Now we proceed to do some webscraping on coronavirus:




```python
from bs4 import BeautifulSoup
import tqdm
import numpy as np
import requests
import re
```

```python
url = 'https://www.worldometers.info/coronavirus/' # This is the url to be scraped from.
response = requests.get(url) # we use this command to get the response from the url
response.text #This command returns a string that contains the information of the url
```

```python
pattern ='"(\w{3} \d{2})"'
dates = re.findall(pattern, response.text) # Using regular expressions to locate date information
dates = dates[:66]
print(dates)
```

```python
dates1= str(dates)
pattern1 = '2020-01-'
pattern2 = '2020-02-'
pattern3 = '2020-03-'
dates2 = re.sub(('Jan '), pattern1, dates1)
dates3 = re.sub('Feb ', pattern2, dates2)
date = re.sub('Mar ', pattern3, dates3)
date = date.replace('\'', '').replace('[', '').replace(']', '').replace(' ', '')
date = date.split(',')
date = pd.to_datetime(date) # Since date was not in datetime format, we applied the "pd.to_datetime" command
date
```
