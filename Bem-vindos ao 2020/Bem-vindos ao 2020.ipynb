## 2. Epidemic Model

```python
from ipywidgets import interact
```

```python
@interact(r = 3.5)
def plot_graph(r):
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    import ipywidgets as widgets

    # SOURCE:
    # https://scipython.com/book/chapter-8-scipy/additional-examples/the-sir-epidemic-model/


    # Total population, N.
    N = 210000000

    # A grid of time points (in days)
    t = range(75)

    # Initial number of infected and recovered individuals, I0 and R0.
    # S0 is the initial number of suscetible. Everyone else is susceptible to infection.
    I0 = 1
    R0 = 0
    S0 = N - I0 - R0

    # R is the `basic reproduction number`: how many people, on average, 1 person infects.
    R = r

    # The mean recovery rate in 1/days
    gamma = 1./14

    # the time between contacts (in days)
    beta = R * gamma

    # The SIR model differential equations.
    from scipy.integrate import odeint
    def sir_rates(y, t, N, beta, gamma):
        """
        S = Susceptible
        I = Infected
        R = Recovered
        """
        S, I, R = y

        # The rate of suscetible people deacreses as the number of infected increases (immunity).
        dSdt = - (beta / N) * S * I 

        # rate of infected is proportional to the number of suscetible ones and the number of infected ones.
        # the coefficient represents how many people one infects in a specific population
        # also, the rate of infected diminishes as a proportion of the rate of recovery increases, hence the second term.
        dIdt = (beta / N) * S * I  - gamma * I

        # rate of recovered is proportional to the number of infected. 
        # this proportion coefficient is the rate of recovery in 1/days
        dRdt = gamma * I

        # returns the rate of susceptible, infected and recovered from infection
        return dSdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, I0, R0

    # Integration of SIR equations over the time grid, t.
    integral = odeint(sir_rates, y0, t, args=(N, beta, gamma))
    S, I, R = integral.T


    # Plot the data on three separate curves for S(t), I(t) and R(t)
    fig, ax = plt.subplots(figsize=(10,7))
    ax.plot(t, I, 'r', alpha=1, lw=3, label='Infected')


    ax.set_xlabel('Time /days')
    ax.set_ylabel('Number (210Ms)')
    ax.set_ylim(0,N*.001)
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid()
    ax.legend()

    for spine in ('top', 'right', 'bottom', 'left'):
        ax.spines[spine].set_visible(False)

    plt.show()
```

```python
import bovespa

bf = bovespa.File('C:\\Users\\Alvaro Camacho\\Desktop\\Ironhack\\Final_Project\\Acao/COTAHIST_A2020.txt')

alimentos = pd.DataFrame()
produtos_farm = pd.DataFrame()
varejo = pd.DataFrame()
combustible = pd.DataFrame()
it = pd.DataFrame()

stock_alimentos = ['BRFS3','CAML3','JBSS3','MDIA3','MRFG3','BEEF3','SMTO3','ABEV3','CRFB3','PCAR3','BKBR3','MEAL3']
stock_produtos_farm = ['NTCO3','HYPE3','RADL3']
stock_varejo = ['ARZZ3','BTOW3','CEAB3','CNTO3','GUAR3','LAME3','LAME4','AMAR3','LREN3','MGLU3','VVAR3','ALPA4','HGTX3',
                'GRND3','VIVA3','VULC3',]
stock_combustible_const = ['CSAN3','PETR3','PETR4','DTEX3',]
stock_it = ['LINX3','POSI3','VIVT4','TIMP3']

def tabela(stock, df):
    """
    Esta função recebe uma lista com os símbolo das empresas e um
    dataframe novo para incluir as informações referentes à ação.    
    """
    for i in stock:
        data = []
        quantidade = []
        acoes = []
        volume = []
        for rec in bf.query(i):
            data.append(rec.date)
            name = rec.company_name
            symbol = rec.stock_code
            acoes.append(rec.price_mean)
            quantidade.append(rec.volume)
            volume.append(rec.quantity)
        df1 = pd.DataFrame({'Data':data,'Açao': name,'Símbolo':symbol,'Preço':acoes, 
                            'Volume':quantidade, 'Quantidade Total':volume})
        df = pd.concat([df,df1], axis=0)
    return df
    
```

```python
produtos_farm = tabela(stock_produtos_farm, produtos_farm)
```

```python
import pandas as pd
from selenium import webdriver
driver = webdriver.Chrome(executable_path='C:\\Users\\Alvaro Camacho\\Desktop\\Selenium/chromedriver.exe')

driver.get('https://www.worldometers.info/coronavirus/country/brazil/')

deaths = driver.execute_script('return window.Highcharts.charts[6].series[0].options.data')
cases = driver.execute_script('return window.Highcharts.charts[2].series[0].options.data')
date = driver.execute_script('return window.Highcharts.charts[2].xAxis[0].categories')
covid = pd.DataFrame(zip(date,cases, deaths))
covid = covid.rename(columns={0:'Data', 1:'Casos de COVID-19', 2:'Mortes Confirmadas'})
covid.Data.replace('Feb ','2020-02-', inplace=True, regex=True)
covid.Data.replace('Mar ','2020-03-', inplace=True, regex=True)
covid.Data.replace('Apr ','2020-04-', inplace=True, regex=True)
covid.Data.replace('May ','2020-05-', inplace=True, regex=True)
covid['Data']= covid.Data.astype(np.datetime64)
covid.to_csv('covid.csv', encoding='latin-1', index=False)
```
